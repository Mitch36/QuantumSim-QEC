<h1> Four Cycle Block, surface code using QuantumSim </h1>
<img src="./pictures/four_cycle_surface_code_white_background.png" alt="Bit flip correction circuit" style="width:50%;"/>
<p> Connection notated as BLUE lines between ancillary qubit measures for BIT flips </>
<p> Connection notated as RED lines between ancillary qubit measures for PHASE flips </>

```python
circuit = Circuit(4)

d1, d2, a1, a2 = 0, 1, 2, 3

# Encode state among two physical qubits
circuit.hadamard(d1)
circuit.cnot(d1, d2)

# Prepare anicallary qubits 
circuit.hadamard(a1)
circuit.hadamard(a2)

# Introduce bit flip on one data qubit
circuit.bitflip_error(d1)

# Introduce phase flip on one data qubit
# circuit.phaseflip_error(d2)

# Syndrome extraction 
circuit.cnot(a1, d1)
circuit.cnot(a1, d2)

# Syndrome extraction - Check for bit flips
circuit.controlled_pauli_z(a2, d1)
circuit.controlled_pauli_z(a2, d2)

# Prepare for measurement 
circuit.hadamard(a1)
circuit.hadamard(a2)

circuit.execute()
circuit.print_gates_and_descriptions()
circuit.measure(print_state=True)
print("A1: " + circuit.get_classical_state_of_qubit_as_string(a1))
print("A2: " + circuit.get_classical_state_of_qubit_as_string(a2))

```

```python
from qiskit import QuantumCircuit
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
 
from qiskit.providers.basic_provider import BasicSimulator
from qiskit.visualization import plot_histogram

# Create a new circuit with two qubits
qc = QuantumCircuit(9, 4)
d1, d2, d3, d4, d5, a1, a2, a3, a4 = 0, 1, 2, 3, 4, 5, 6, 7, 8

qc.h(d1)
qc.cx(d1, d2)
qc.cx(d2, d3)
qc.cx(d3, d4)
qc.cx(d4, d5)

# Introduce errors
qc.x(d1)
qc.z(d5)

# Measure bit flips (blue lines)
qc.cx(d1, a2)
qc.cx(d3, a2)
qc.cx(d4, a2)
qc.cx(d2, a3)
qc.cx(d3, a3)
qc.cx(d5, a3)
qc.h(a1)
qc.h(a4)
qc.cx(a1, d1)
qc.cx(a1, d2)
qc.cx(a1, d3)
qc.cx(a4, d3)
qc.cx(a4, d4)
qc.cx(a4, d5)
qc.h(a1)
qc.h(a4)

qc.measure(a1, 0)
qc.measure(a2, 1)
qc.measure(a3, 2)
qc.measure(a4, 3)



# Return a drawing of the circuit using MatPlotLib ("mpl"). This is the
# last line of the cell, so the drawing appears in the cell output.
# Remove the "mpl" argument to get a text drawing.
qc.draw("mpl")

backend = BasicSimulator()
result = backend.run(qc, shots=100).result()   
counts = result.get_counts()                    
plot_histogram(counts)
```

```python
def bloch_to_probabilities(bloch_vector):
  """
  Calculates the probabilities of measuring a qubit in the |0 or |1 state,
  given its Bloch sphere coordinates.

  Args:
    bloch_vector: A tuple or list of three floats representing the (x, y, z) coordinates
                  of the Bloch vector.

  Returns:
    A tuple of two floats representing the probabilities of measuring |0 and |1.
  """

  x, y, z = bloch_vector

  # Calculate spherical coordinates
  theta = np.arccos(z)
  phi = np.arctan2(y, x)

  # Calculate probability amplitudes
  # This is almost the general matrix for one qubit 
  alpha = np.cos(theta / 2)
  beta = np.sin(theta / 2) * np.exp(1j * phi)

  # Calculate probabilities
  p0 = np.abs(alpha)**2
  p1 = np.abs(beta)**2

  return p0, p1


circuit = Circuit(1, 1, True)

circuit.pauli_x(0)

# 1. Kopieer state vector
copyStateVector = circuit.state_vector
print("copyStateVector: " + copyStateVector.measure())

# 2. Meet gekopieerde state vector
if(copyStateVector.get_classical_state_as_string() == "|1>"):
  # 3. Reset actuele state vector om de state |0> te krijgen, measurement mimic
  print("Reset with Pauli x")
  circuit.pauli_x(0)

# circuit.hadamard(0)
print(circuit.measure())

circuit.state_vector.print()

# The measurement gate can be implemented by measuring all qubits and running the circuit again but than without the measurement blocks
# Assumption is that there are no random variables and the same noise can be reapplied. 

# State vector kopieren en daarmee verder rekenen, dat kan niet sinds je na measurement de qubit in de basis states moet plaatsen.


```
