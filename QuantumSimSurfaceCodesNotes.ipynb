Example with Two z stabilizers and one x stabilizer among four data qubits
==========================

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
 
from qiskit.providers.basic_provider import BasicSimulator
from qiskit.visualization import plot_histogram

qrd = QuantumRegister(4, "Data")
qra = QuantumRegister(3, "Ancillary")
cr = ClassicalRegister(3,"Syndrome")
qc = QuantumCircuit(qrd, qra, cr)
d1, d2, d3, d4, x1, z1, z2 = 0, 1, 2, 3, 4, 5, 6

qc.h(d1)
qc.cx(d1, d2)
qc.cx(d2, d3)
qc.cx(d3, d4)

qc.barrier()

qc.z(d2)
qc.x(d2)

qc.h(z1)
qc.h(z2)
qc.h(x1)

qc.cx(x1, d2)
qc.cz(z1, d1)


qc.cx(x1, d1)
qc.cz(z1, d3)


qc.cx(x1, d3)
qc.cz(z2, d2)

qc.cx(x1, d4)
qc.cz(z2, d4)

qc.barrier()


qc.h(z1)
qc.h(z2)
qc.h(x1)

qc.measure(x1, 0)
qc.measure(z1, 1)
qc.measure(z2, 2)

# qc.draw("mpl")

backend = BasicSimulator()
result = backend.run(qc, shots=100).result()   
counts = result.get_counts()                    
plot_histogram(counts)
```

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
 
from qiskit.providers.basic_provider import BasicSimulator
from qiskit.visualization import plot_histogram

qrd = QuantumRegister(6, "Data")
qra = QuantumRegister(2, "Ancillary")
cr = ClassicalRegister(2,"Syndrome")
qc = QuantumCircuit(qrd, qra, cr)
d1, d2, d3, d4, d5, d6, x1, z1 = 0, 1, 2, 3, 4, 5, 6, 7

# https://errorcorrectionzoo.org/c/surface-17#citation-3
# https://arxiv.org/pdf/2303.17211

qc.h(d2)
qc.h(d4)
qc.h(d6)

qc.cx(d2, d1)
qc.cx(d4, d5)
qc.cx(d4, d3)
qc.cx(d6, d5)

qc.x(d1)
qc.z(d4)

qc.barrier()

qc.cx(d3, d2)

qc.h(x1)
qc.h(z1)

qc.cx(x1, d4)

qc.cx(x1, d5)

qc.cx(x1, d2)

qc.cx(x1, d3)

qc.cz(z1, d1)
qc.cz(z1, d2)
qc.cz(z1, d5)
qc.cz(z1, d6)

qc.barrier()

qc.h(x1)
qc.h(z1)

qc.measure(x1, 0)
qc.measure(z1, 1)

# qc.draw("mpl")

backend = BasicSimulator()
result = backend.run(qc, shots=100).result()   
counts = result.get_counts()                    
plot_histogram(counts)
```
