In this Jupyter notebook we show how intermediate probabilities and states can be visualised using functions *show_all_probabilities()* and show_all_intermediate_states()*.

Import the QuantumSim code.

```python
from quantumsim import *
```

### Some examples

```python
circuit = Circuit(4)
circuit.hadamard(0)
circuit.swap(0,1)
circuit.swap(1,2)
circuit.swap(2,3)
result = QuantumUtil.run_circuit(circuit)
QuantumUtil.histogram_of_classical_states(result)
QuantumUtil.show_all_pobrabilities(circuit)
QuantumUtil.show_all_intermediate_states(circuit)
```

```python
circuit = Circuit(4)
circuit.pauli_x(0)
circuit.pauli_y(0)
circuit.hadamard(0)
circuit.cnot(0,1)
circuit.cnot(1,2)
circuit.cnot(2,3)
result = QuantumUtil.run_circuit(circuit)
QuantumUtil.histogram_of_classical_states(result)
QuantumUtil.show_all_pobrabilities(circuit)
QuantumUtil.show_all_intermediate_states(circuit)
```

```python
circuit = Circuit(4)
circuit.pauli_x(0)
circuit.cnot(0,1)
circuit.controlled_pauli_y(0,2)
circuit.controlled_pauli_z(0,3)
result = QuantumUtil.run_circuit(circuit)
QuantumUtil.histogram_of_classical_states(result)
QuantumUtil.show_all_pobrabilities(circuit)
QuantumUtil.show_all_intermediate_states(circuit)
```

```python
circuit = Circuit(4)
circuit.hadamard(0)
circuit.controlled_hadamard(0,1)
circuit.controlled_hadamard(1,2)
circuit.controlled_hadamard(2,3)
result = QuantumUtil.run_circuit(circuit)
QuantumUtil.histogram_of_classical_states(result)
QuantumUtil.show_all_pobrabilities(circuit)
QuantumUtil.show_all_intermediate_states(circuit)
```

```python
circuit = Circuit(4)
for q in range(4):
    circuit.pauli_x(q)
circuit.hadamard(0)
circuit.controlled_hadamard(0,1)
circuit.controlled_hadamard(1,2)
circuit.controlled_hadamard(2,3)
result = QuantumUtil.run_circuit(circuit,2000)
QuantumUtil.histogram_of_classical_states(result)
QuantumUtil.show_all_pobrabilities(circuit)
QuantumUtil.show_all_intermediate_states(circuit)
```

```python
circuit = Circuit(4)
theta = np.pi/8
for q in range(4):
    circuit.pauli_x(q)
    circuit.phase(q*theta, q)
result = QuantumUtil.run_circuit(circuit)
QuantumUtil.histogram_of_classical_states(result)
QuantumUtil.show_all_pobrabilities(circuit)
QuantumUtil.show_all_intermediate_states(circuit)
```

```python
circuit = Circuit(1)
circuit.pauli_x(0)
theta = np.pi/8
for q in range(16):
    circuit.phase(theta, 0)
result = QuantumUtil.run_circuit(circuit)
QuantumUtil.histogram_of_classical_states(result)
QuantumUtil.show_all_pobrabilities(circuit)
QuantumUtil.show_all_intermediate_states(circuit)
```
